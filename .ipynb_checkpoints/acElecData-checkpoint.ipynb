# PSOE and PP Autonomous Community election result analysis from 1978 to 2020

This data analysis provides a basic analysis of the election results of Spain's two major national parties, the Partido Socialista Obrero Español (PSOE) and the Partido Popular (PP), at the Autonomous Community (AC) level from 1978 to 2020. The period from 1978 to 2020 can be divided into roughly three periods, a period of political dominance of the PSOE from 1978 to 1996, followed by a period of realtively even competition between the PP and PSOE and which includes massive important electoral setbacks for each, followed by a "new era" of Spanish politics from 2015. The aim of the subsequent analysis is to provide an overview of the performance of the POSE and PP at the sub-national Autonomous Community level. The following section will provide a more detailed discussion of these periods and the subsequent section will provide a brief discussion of key findings.        
## National background to AC elections
At the national level, both parties have experienced important periods of electoral growth and decline. Generally speaking, the massive decline of the Unión de Centro Democrático (UCD) at the national elections of 1982 permitted the PSOE to become the lagest party in parliament, followed by the Alianza Popular (AP), the predecessor of today's PP. From 1982 to 1989, the PSOE enjoyed a clear electoral dominance over the PP, but the PP managed to expand its share of the vote from 1989 and won the national elections for the first time in 1996, thus inaugarating a period in which the PSOE and PP competed with one another as more or less equally strong contenders until 2011. Following the PSOE's victory in 2008, the outbreak of the global financial crisis that year led to a significant loss of electoral support for the PSOE and returned the PP to government in 2011 with an absolute majority. Nonetheless, by 2015 the PP too had become increasingly unpopular, leading to the breakthrough of two new national parties, the far left Podemos and the center right Ciudadanos. Both were significant political parties until 2019, after which especially Ciudadanos has experienced setbacks which have reduced it to a party of secondary political importance. The following section will provide the most important findings of the analysis. 

## Data analysis: 

```python
%matplotlib inline 
%config InlineBackend.figure_format = "svg" #better resolution of plots
from IPython.core.display import HTML #center cell ouput
display(HTML("""
<style>
.output {
    display: flex;
    align-items: center;
    text-align: center;
}
</style>
"""))

import warnings #ignore/hide warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
pd.set_option('display.precision', 1) #Rounding numbers in tables
import seaborn as sns
from matplotlib import pyplot as plt

plt.rcParams.keys()
plt.rcParams.update({"font.size": 12, 
                     "font.family": "Times New Roman",
                     "figure.titlesize": 14.5,
                     "axes.titlesize":12, #changes size of titles of plots in figure-level plot
#                     'axes.labelsize': 'medium',
#                     "figure.figsize" : [2.8,2.4],
                     "figure.dpi": 150
                    })
#fontsizefloat or {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}

df = pd.read_pickle("acElecDataClean.pkl")

colPal = ["#FF0000", "#0000FF"]
colOrder = ["PSOE", "PP"]
yearParams = [[1978, 2022], [1978, 1992], [1993, 2010], [2011, 2014], [2015, 2018], [2019, 2021]]
```

The boxplot below shows the distribution of vote results across the different communities by party. A number of general observations can be made. Firstly, taking the median as a point of orientation, the PSOE's performance has been worst in the Basque Country, Navarra, and Galicia. The PP's performance has been worst in Catalonia and the Basque Country and Canarias. It is noticable that the PP's performance in Catalonia and the Basque Country is worse by a clear margin both when compared to the PP's electoral results in other ACs and when compared to the PSOE's worst results. The PSOE's performance has been best in Extremadura, Castilla-La Mancha and Andalusia, where the medians lie above 45 percent. The PP has done particularly well in Madrid, Galicia and Castilla-y-Leon, although its performance has also been similarly positive in La Rioja. Although Murcia and the Balearic Islands have a similar median, their spread is much greater and are thus not considered too similar. That the PP's predecessor's party, the AP, was initially quite unpopular is also reflected in its comparatively larger interquartile ranges in a number of ACs. Particularly noticeable in this sense are Navarra, Valencia, Murcia and Madrid. The PSOE's interquartile range is especially large in Cantabria and Catalonia. There are also a number of outliers: **XXX**

```python
### Boxplot, PSOE and PP by AC across time (boxplot)
majorParties = df[(df["party"].isin(["PSOE", "PP"]))].sort_values("AC")
orderPlot = majorParties.groupby(["party", "AC"]).votePerc.median().reset_index().sort_values(["party", "votePerc"], ascending = [False, True])
orderPlot.set_index(orderPlot.party+orderPlot.AC, inplace = True)
orderPlot.index.values

majorParties.set_index(majorParties.party+majorParties.AC, inplace = True)
majorParties["majorPartiesCats"] = majorParties.index.values
plotBox = sns.catplot(x = "votePerc", 
            y = "majorPartiesCats", 
            kind = "box", 
            palette = colPal,
            hue = "party", 
            hue_order = ["PSOE", "PP"],
            order = orderPlot.index.values,
            height=6.5, 
            aspect=1.2, 
            data = majorParties,
            facet_kws={'legend_out': False})
sns.move_legend(
    plotBox, "lower center",
    bbox_to_anchor=(.5,.965),
    ncol=2,
    title=None,
    frameon=False,
    fontsize = 12)
# plotBox._legend.set_title("Party")
plotBox.set_yticklabels(majorParties.groupby(["party", "AC"]).votePerc.median().reset_index().sort_values(["party", "votePerc"], ascending = [False, True])["AC"])
plt.xlabel("Vote percentage")
plt.ylabel("Autonomous Community")
plotBox.fig.suptitle("Election results by AC across time", y = 1.06, size = 14.5);
#https://stackoverflow.com/questions/36220829/fine-control-over-the-font-size-in-seaborn-plots
```

```python
### Main parties across time periods by AC
majorParties.reset_index(drop = True, inplace = True) 
def partySysPhases(yearStartEnd, df = majorParties, y = "votePerc", titleInsert = "election results",
                   hline = False):
    """yearStartEnd is list containing integer of start and integer of end of period """
    yearStart = yearStartEnd[0]
    yearEnd = yearStartEnd[1]
    df = df[df["year"].isin([i for i in range(yearStart, yearEnd)])].sort_values("AC")
    fig1 = sns.relplot(x= "date", y = y, 
                       hue="party", 
                       palette = colPal, 
                       hue_order = colOrder, 
                       kind="line", 
                       linewidth = 2.75,
                       height = 2.8,
                       aspect = 2.4/2.8,
                       marker = "o", 
                       col = "AC", 
                       col_wrap = 3, 
                       data = df,
                       facet_kws={'legend_out':False})
    fig1._legend.set_title("Party")
    sns.move_legend(fig1, "upper left", bbox_to_anchor=(.75, .13))
    fig1.fig.suptitle("PP and PSOE "+titleInsert+", "+str(yearStart)+"-"+str(yearEnd), y = 1.015, size = 14.5)
    fig1.set_axis_labels("Year", "Vote percentage")    
    axes = fig1.axes.flatten()
    for i, name in enumerate(df.AC.unique()):
        axes[i].set_title(name)
    for ax in axes:
        ax.tick_params(axis = "x", labelbottom=True, labelsize = 10, labelrotation = 45)
        if hline == True:
            ax.axhline(0, linestyle = "--", color = "black")
    plt.subplots_adjust(wspace = 0.1, hspace=0.3)
    plt.show(fig1)
    return fig1
g = partySysPhases(yearParams[0]);
```

The next series of plots show the electoral results of both parties across all ACs. It becomes evident that AC electoral trends reflect national ones, although this is best investigated via correlation or regression (see forthcoming work on this topic). Summarising the patterns visible in the above plots:
- In the majority of ACs, the PSOE was the initially dominant party, with the exception of Cantabria, Galicia and Balearic Islands, where the PP was more popular from the start and in which the PP maintained a strong dominance with only one exception throughout (and despite heavy losses in the former and the latter).
- There are a number of ACs which became highly contested between the two parties during the 1990s and 2000s, which include Aragon, Asturias and Canarias.
- There are a number of ACs in which the PP established dominance during the 1990s and 2000s, including Cantabria, Castilla-y-Leon, La Rioja, Madrid, Murcia, Navarra and Valencia.
- There are a number of ACs in which the PSOE could roughly hold onto its electoral dominance until 2011, namely Andalusia, Castilla-La Mancha, Catalonia, Extremadura. The PP enjoyed dominance from the start in a number of 
- The PSOE, starting from a high degree of popularity in most ACs, experinced important setbacks around 1995, thereafter peaking again in the majority of ACs around 2010, after which a massive and continous decline set in, which was only very partially reversed in 2020. 
- The PP, from a lower starting point, increased in popularity until pealing or plateuaing in most ACs around 1995. It experienced another peak in popularity around 2010. Thereafter, however, the PP also suffered massive electoral declines, which continued until 2020.  
The general trend, with noticable exceptions, thus reflects that AC electoral results clearly preceed or follow the national trend, with peaks and *throughs* of popularity clearly complementing each other until 2015, when two new national contestors experienced their breakthrough and were capable of absorbing the disatisfaction of the PP and PSOE without benefiting either one. The following plot summarizes these developments with a rolling average (shown in black), necessary since some ACs, especially Catalonia, the Basque Country and Andalusia hold elections at other points in time than the rest of ACs.

```python
### Diff in percentage with rolling average, one plot for trend
majorParties = majorParties.sort_values(["AC", "party", "year", "month"])
majorParties["votePercLagged"] = majorParties.groupby(["AC", "party"])["votePerc"].shift(1)
majorParties["votePercDiff"] = majorParties.votePerc - majorParties.votePercLagged

#17 comunidades, 13 normalmente votan juntos
majorParties2 = majorParties
majorParties2['rollingVotePerc'] = majorParties.sort_values(["year", "month"]).groupby("party")["votePerc"].transform(lambda x: x.rolling(13).mean())
majorParties2 = majorParties2[["rollingVotePerc", "party", "date"]].rename({"rollingVotePerc":"votePerc",}, axis = "columns")
majorParties2["AC"] = "yRollingAvg"
majorPartiesRoll = pd.concat([majorParties, majorParties2]).reset_index(drop=True)

fig2 = sns.relplot(x = "date", y = "votePerc", 
                   col ="party", 
                   col_order = colOrder,
                   hue = "AC",
                   height = 3.6,
                   aspect = 1,
                   palette = ["gray"]*(len(majorPartiesRoll.AC.unique())-1)+["black"],
                   kind = "line",
                   legend = False,
                   data = majorPartiesRoll) 
fig2.fig.suptitle("Overview of vote evolution across time", y = 1.05, size = 14.5) # Title size
fig2.set_axis_labels("Year", "Vote difference (percent)")
plt.subplots_adjust(wspace = 0.05)
#plt.ylim(reversed(plt.ylim()))
axes = fig2.axes.flatten()
for i, name in enumerate(colOrder):
    axes[i].set_title(name)
for i, axis in enumerate(axes): 
    axes[i].tick_params(axis = "x", labelbottom=True, labelsize = 10.5, labelrotation = 45);
```

## The dominance of the PSOE and PP over the different party system periods
The following heatmap plots provide a quick visual overview over the average electoral results by the PSOE and PP across the different phases of the Spanish electoral system. 

As can be seen in the plot below, the first period of AC elections clearly reflected the PSOE's dominance. In nine communities, the PSOE achieved higher average vote results than the PP achieved in any AC during that same period. Especially noticeable are Extremadura, Andalucia, Castilla-La Mancha, Murcia and Valencia, where the PSOE won more than 45 percent of the vote on average. The only ACs where the PP led by a clear margin are Castilla-y-Leon, Balearic Islands and Galicia. The PSOE achieved its worst results in the most nationalist regions, namley Catalonia, Basque Country, Galicia and Navarra, ranging between 29.2 and 19.8 percent. The PP performed especially bad in Calatonia and the Basque Country, winning only an average 6.5 and 6.8 percent of the vote. In Aragon, Navarra and Canary Islands, the PP's results ranged from a mean 19.8 to a mean 17.6 percent. 

```python
### Heatmaps across time
def heatMap(yearStartEnd, df = majorParties):  
    yearStart = yearStartEnd[0]
    yearEnd = yearStartEnd[1]
    df = df[df["year"].isin([i for i in range(yearStart, yearEnd)])]
    df = df[["AC", "party", "votePerc"]].pivot_table(columns="party", index="AC", values ="votePerc").iloc[:,[1,0]].sort_values("PSOE", ascending = False)
    df = df.astype("float")
    plt.figure(figsize = (6.5,6.5*1.2))
    heatMap = sns.heatmap(df, annot=True, fmt=".1f", cmap ="RdYlGn")
    plt.xlabel("Party")
    plt.ylabel("Autonomous Community")
    heatMap.set_title("Average election results, " + str(yearStartEnd[0]) + "-" + str(yearStartEnd[1]), y = 1.025, size = 14.5)
    return heatMap
out = heatMap(yearParams[1])    
```

The following period from 1993 to 2010, visualized below, brought significant changes. The PP achieved an impressive mean of 55.7 percent in Murcia and between 50.3 and 51.4 percent of the mean vote in Galicia, La Rioja, Madrid and Castilla-y-Leon. Except for Catalonia, where the party's performance continued to be dismal, having increased to only 11.4 percent, it achieved significant increases where it had previously been rather weak, the most embelemetic case being the Basque Country, where it achieved a mean of 17.8 percent, Canarias, were it rose to a mean of 28.6, and Navara, where it reached 39.9 percent. When one compares the number of ACs where each party scored a mean above 40 percent, this change becomes all to evident: the PP scored above 39 percent in 12 ACs, the PSOE only in four, namely Castilla-La-Mancha, Extremadura, Andalucia and Asturias.

## Vote change across time
Another way to analyse these changes is to look at vote percentage differences from one election to the next. The graph below, which does this for the entire 1978 to 2022 period, shows that, at least in general, when one party makes large gains in one community, this is usually accompanied by large losses of the other. There are some important exceptions to this general trend, such as in Cantabria and in Asturias and Navarra in 2011, all owed to regionalist and nationalist parties. Even in the Basque Country and Catalonia, however, where nationalist parties are of great importance, this patterns holds. 

```python
### Diff in percentage of vote across time by AC
partySysPhases(yearParams[0], df = majorParties, y = "votePercDiff", hline = True, titleInsert = "vote percentage changes");
```

A better overview in vote change can be obtained by considering the following graph. It shows mean vote differences for the periods of the Spanish party system, slightly adjusted to better suit the subnational electoral trends identified above. On average, from 1978 to 1992, the PSOE experienced slight average losses (-0.85 percent) and the PP slight gains (0.22) close to zero percent. In the 1993 to 2010 period, the PSOE, which experienced both losses around 1995 and a recovery around 2004, again largely maintaining its mean losses near zero (-0.55), while the PP experiences average gains of 2.7 percent. The period from 2011 to 2015, however, is marked by massive average declines for both parties, which each fall by 6.4 percent. The PP's standard deviation is particularly large in this period, more than twice that of the PSOE, given the party's gains around 2011 and massive losses in 2015, while PSOE declined throughout. Between 2016 and 2022, however, the PP declined another 3.7 percent, while the PSOE began to recover by 2.6 percent on average. **REWRITE WITH DIFF PERIOD** 

```python
### Means and std for different periods
period = []
for year in majorParties["year"]:
    if year < 1993:
        period.append("1978-92")
    elif year < 2011:
        period.append("1993-2010")
    elif year < 2014:
        period.append("2011-14")
    elif year < 2019:
        period.append("2015-18")
    else:
        period.append("2019-21")
majorParties["period"] = period
majorParties["period"] = majorParties.period.astype("category")

stats = majorParties.groupby(["period", "party"])[["votePercDiff"]].agg(["mean", "std"]).round(2).reset_index()
fig3 = sns.catplot(x = "period", y= "votePercDiff", hue = "party", kind = "point", 
            palette = colPal,
            hue_order = colOrder,
            dodge = True,
            height = 3.6,
            aspect = 1.1,
            capsize = .3,
            data = majorParties,
            facet_kws={'legend_out': False})
fig3.axes.flatten()[0].tick_params(axis = "x", labelbottom=True, labelsize = 11)
fig3.set_axis_labels("Period", "Vote difference (percent)")
sns.move_legend(
    fig3, "lower center",
    bbox_to_anchor=(.5,.9),
    ncol=2,
    title=None,
    frameon=False,
    fontsize = 12)
fig3.fig.suptitle("Mean vote change across different periods", y = 1.05, size = 14.5);
```

```python
### Absolute majorities
absMaj = majorParties[majorParties["depsPerc"] >= 50].groupby(["party", "AC"])[["AC"]].count().rename({"AC":"Count"}, axis = 1)
absMaj = absMaj[absMaj["Count"]!=0].sort_values(["party", "Count"], ascending = False)
# also by period
absMajPeriod = majorParties[majorParties["depsPerc"] >= 50].groupby(["party", "period", "AC"])[["AC"]].count().rename({"AC":"Count"}, axis = 1)
absMajPeriod = absMajPeriod[absMajPeriod["Count"]!=0].sort_values(["party", "period", "Count"], ascending = [True, False, False]).reset_index()
absMajAgg = absMajPeriod.groupby(["period", "party"]).sum("Count").reset_index()

fig4 = sns.relplot(x= "period", 
            y= "Count", 
            hue="party", 
            palette = colPal, 
            hue_order = colOrder, 
            linewidth = 2.75, 
            height = 3.6,
            aspect = 1.1,
            marker = "o",  
            kind="line", 
            data = absMajAgg,
            facet_kws={'legend_out': False})
sns.move_legend(
    fig4, "lower center",
    bbox_to_anchor=(.5,.92),
    ncol=2,
    title=None,
    frameon=False,)
fig4.axes.flatten()[0].tick_params(axis = "x", labelbottom=True, labelsize = 11)
fig4.set_axis_labels("Period")
fig4.fig.suptitle("Number of absolute majorities", y = 1.075, size = 14.5);
print(absMajAgg)
```

## The drastic loss of regional power and conclusion

These drastic losses of both major parties have led to a significant loss of political power, as is best reflected in the massive changes in absolute majority of seats held by the major parties. Whereas from 1978 to 1992, the PP only boasted four ansolute majorities, it acheived 24 absolute majorities in the subsequent period from 1993 to 2010. The PSOE, on the other hand, experience a decline from 19 to 10 absolute majorities during the same time. From 2011 to 2018, the PSOE won no absolute majority and only won one in the 2019 to 2021 period. The PP experienced a similar, albeit delayed, fait. After dropping from 14 to 9 absolute majorities by the end of 2014, that number declined to two in the 2015 to 2018 period and to none thereafter until 2021. 

Overall, then, this analysis has shown that AC election outcomes tend to follow, with some noticeable exceptions, national political trends. The different points of departure for both parties nonetheless resulted in a somewhat contrary initial evolution. While the PSOE continued to lose vote percentages from 1978, except for a slight in the 2000s, the PP grew stronger and stronger until roughly 2015, when it suffered a series of equally devastating defeats. as of 2019 to 2021, the PSOE has recovered somewhat, but compared to the losses of the past, the party is still a long way from recovering its former dominance. The PP, on the other hands continued its decline as of 2021.
